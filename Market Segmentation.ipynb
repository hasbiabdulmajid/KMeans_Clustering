# Market Segmentation Using K-Means Clustering

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load Data

```python
data = pd.read_csv('3.12. Example.csv')
data
```

## Plot Data

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

## Select Feature

```python
x = data.copy()
```

## Clustering

```python
model = KMeans(2)
model.fit(x)
```

```python
clusters = x.copy()
clusters['Cluster Prediction']= model.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'],clusters['Loyalty'],c=clusters['Cluster Prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

## Standarize the feature

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

## Elbow Method

```python
wcss = []
wcss2 = []
for i in range(1,10) :
    # Scaled Features
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
    #Unscaled Features
    kmeans2 = KMeans(i)
    kmeans2.fit(x)
    wcss2.append(kmeans2.inertia_)
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.title('Elbow Method of Scalled Feature')
plt.xlabel('Number of Cluster')
plt.ylabel('WCSS')
```

```python
plt.plot(range(1,10),wcss2)
plt.title('Elbow Method of Unscalled Feature')
plt.xlabel('Number of Cluster')
plt.ylabel('WCSS')
```

## Exploring Clustering Solution and Select the Number of Cluster (K)

### K = 2

```python
model2 = KMeans(2)
model2.fit(x_scaled)
cluster_new = x.copy()
cluster_new['Cluster Prediction'] = model2.fit_predict(x_scaled)
```

```python
plt.scatter(cluster_new['Satisfaction'],cluster_new['Loyalty'],c=cluster_new['Cluster Prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

### K = 3

```python
model3 = KMeans(3)
model3.fit(x_scaled)
cluster3 = x.copy()
cluster3['Cluster Prediction'] = model3.fit_predict(x_scaled)

# Plot 
plt.scatter(cluster3['Satisfaction'],cluster3['Loyalty'],c=cluster3['Cluster Prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

### K = 4

```python
model4 = KMeans(4)
model4.fit(x_scaled)
cluster4 = x.copy()
cluster4['Cluster Prediction'] = model4.fit_predict(x_scaled)

# Plot 
plt.scatter(cluster4['Satisfaction'],cluster4['Loyalty'],c=cluster4['Cluster Prediction'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```
